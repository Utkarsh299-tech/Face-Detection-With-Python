```
import numpy as np
import matplotlib.pyplot as plt
```

```
img = plt.imread("/content/group_photo.jpg")
```

```
plt.imshow(img)
```

```
import cv2
```

```
model = cv2.CascadeClassifier("/content/haarcascade_frontalface_default.xml")
```

```
faces = model.detectMultiScale(img)
```

```
faces
```

```
faces.shape
```

```
for face in faces:
  x,y,w,h = face

  img = cv2.rectangle(img, (x,y), (x+w, y+h), (0,0,0), 2)
```

```
plt.imshow(img)
```
